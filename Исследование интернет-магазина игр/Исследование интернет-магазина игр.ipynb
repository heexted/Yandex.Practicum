# Исследование продаж интернет-магазина "Стримчик"

Я работаю в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). **Нужно выявить определяющие успешность игры закономерности.** Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

Передо мной данные до 2016 года. Представим, что сейчас декабрь 2016 г., и я планирую кампанию на 2017-й. Нужно отработать принцип работы с данными.

ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

```python
#импорт необходимых библиотек
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
import numpy as np
from scipy import stats as st
```

### Подготовка данных

#### Замена названия столбцов

```python
#переименовывание столбцов к нижнему регистру
data.columns = data.columns.str.lower()
data.head(1)
```

#### Преобразование данных в нужные типы и замена пропусков

```python
#удаление пропусков с годом релиза
data = data.dropna(subset=['year_of_release']) 
#изменение типа данных
data['year_of_release'] = data['year_of_release'].astype('int64')
data['year_of_release'].isna().sum()
```

```python
#замена 'tbd' на nan
data.loc[data['user_score'] == 'tbd', 'user_score'] = float('nan')
#замена пропущенных знаений на 0
data['user_score'] = data['user_score'].fillna(0)
#изменение типа данных
data['user_score'] = data['user_score'].astype('float64')
data['user_score'].isna().sum()
```

```python
#замена пропущенных значений на 'unknown'
data['rating'] = data['rating'].fillna('unknown')
```

```python
#замена пропущенных знаений на 0
data['critic_score'] = data['critic_score'].fillna(0)
```

```python
#проверка
data.isna().sum()
```

```python
#удаляем оставшееся пропуски
data = data.dropna(subset=['name']) 
data.isna().sum()
```

Пропущенные данные годов релиза были удалены, т.к. составляли около 2 %, оставшееся значения изменили на тип 'int64'

Пропущенные данные пользовательских оценок и 'tbd' были замены на 0. Оценки критиков так же были заменены на 0. 
Пропущенные значения рейтинга от организации ESRB заменили на заглушки 'unknown'.
**Любое заполнение этих пропусков исказит результаты корреляционного анализа и стат-тестов.**

Возможные причины возниктовения пропусков:
* год релиза - скорее всего, тех. ошибка при записи данных
* оценки критиков и пользователей - либо недостаточное кол-во оценок, поэтому данные не прогрузились, или тех. ошибка
* рейтинг ESRB - на каждой игре обязательно указан возрастной рейтинг, возможно, ошибки при записи данных

```python
#создание нового столбца суммы продаж
data['total_sales'] = data[['na_sales','eu_sales','jp_sales', 'other_sales']].sum(axis = 1)
data.head()
```

После построения распределения по годам можно заметить, что в среднем старые платформы исчезают в период с 2011 - 2016 года. Новые платформы появляюся с 2011-2013. В среднем платформы сущетвуют до 8 лет.

#### Зависимость объема продаж внутри одной популярной платформы от отзывов пользователей и критиков

```python
#рассмотрим зависимость продаж от отзывов критиков
data_actual_PS4 = data_actual.query('platform == "PS4"')
data_actual_PS4.plot(
    x='critic_score',
    y='total_sales',
    kind='scatter',
    alpha=0.7,
    title='Зависимость продаж от отзывов критиков',
    figsize=(7,5),
    xlabel='Отзывы критиков',
    ylabel='Объем продаж, млн',
    grid=True);
print('Корреляция Пирсона составляет:', data_actual_PS4['total_sales'].corr(data_actual_PS4['critic_score']))
```

```python
#рассмотрим зависимость продаж от отзывов пользователей
data_actual_PS4.plot(
    x='user_score',
    y='total_sales',
    kind='scatter',
    alpha=0.7,
    title='Зависимость цены от отзывов пользователей',
    figsize=(7,5),
    xlabel='Отзывы пользователей',
    ylabel='Объем продаж, млн',
    grid=True);
print('Корреляция Пирсона составляет:', data_actual_PS4['total_sales'].corr(data_actual_PS4['user_score']))
```

Можно сделать вывод, что зависимость между отзывами критиков и продажами небольшая, но есть. Чем выше оценки, тем больше объем продаж.

Оценки пользователей почти не влияют на объем продаж, чем оценки пользователей.

#### Сравнение зависимости продаж игр от отзывов на других платформах

```python
#рассмотрим зависимость продаж от отзывов критиков
for platform in data_actual['platform'].unique():
    data_actual.plot(
    x='critic_score',
    y='total_sales',
    kind='scatter',
    alpha=0.7,
    title=platform,
    figsize=(7,5),
    xlabel='Отзывы критиков',
    ylabel='Объем продаж, млн',
    grid=True);
    print(
        'Корреляция Пирсона платформы',  platform, 'составляет:', 
        data_actual['total_sales'].loc[data_actual['platform'] == platform].corr(
            data_actual['critic_score'].loc[data_actual['platform'] == platform]))
```

```python
#рассмотрим зависимость продаж от отзывов пользователей
for platform in data_actual['platform'].unique():
    data_actual.plot(
    x='user_score',
    y='total_sales',
    kind='scatter',
    alpha=0.7,
    title=platform,
    figsize=(7,5),
    xlabel='Отзывы пользователей',
    ylabel='Объем продаж, млн',
    grid=True);
    print(
        'Корреляция Пирсона платформы',  platform, 'составляет:', 
        data_actual['total_sales'].loc[data_actual['platform'] == platform].corr(
            data_actual['user_score'].loc[data_actual['platform'] == platform]))
```

Диаграммы рассеянивания подтверждают предыдущий вывод, что оценки критиков влияют на объем продаж. А оценки пользователей почти нет.

### Проверка гипотез

В Python есть специальный метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам t-test. Для проверки будем использовать метод двухсторонней оценки scipy.stats.ttest_ind.

Для проведения теста должы быть соблюдены условия :

* Генеральные совокупности не должны зависеть друг от друга.

* Выборочные средние должны быть нормально распределены.

* Дисперсии рассматриваемых генеральных совокупностей должны быть равны. 

* Если выборки разного размера или есть сомнения, что дисперсии у совокупностей одинаковые, укажите этот параметр(equal_var) как False, и тест посчитает дисперсии выборок по отдельности.


#### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

```python
xbox = data_actual.loc[data_actual['platform'] == 'XOne'].query('user_score != 0')['user_score']
pc = data_actual.loc[data_actual['platform'] =='PC'].query('user_score != 0')['user_score']
print('размер выборки action', len(xbox))
print('paзмер выборки sports', len(pc))
```

```python
xbox.hist(bins=25)
```

```python
pc.hist(bins=25)
```

Нулевая гипотеза - средние рейтинги платфоры Xbox One = средние рейтинги платфоры PC

Альтернативная гипотеза - средние рейтинги платфоры Xbox One ≠ средние рейтинги платфоры PC 

Размеры выборок разные, средние распределены нормально.


```python
alpha = .05 #критический уровень статистической значимости
results = st.ttest_ind(xbox, pc, equal_var=False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
    
print('Средний рейтинг Xbox', xbox.mean())
print('Средний рейтинг PC', pc.mean())
```

Вывод: средние рейтинги Xbox и PC дейстительно равны.

#### Средние пользовательские рейтинги жанров Action и Sports разные.

```python
action = data_actual.loc[data_actual['genre'] =='Action'].query('user_score != 0')['user_score']
sports = data_actual.loc[data_actual['genre'] == 'Sports'].query('user_score != 0')['user_score']
print('размер выборки action', len(action))
print('paзмер выборки sports', len(sports))
```

```python
action.hist(bins=25)
```

```python
sports.hist(bins=25)
```

Нулевая гипотеза - средние рейтинги жанра Action =  средние рейтинги жанра Sports

Альтернативная гипотеза - средние рейтинги жанра Action ≠ средние рейтинги жанра Sports

Размеры выборок разные, распередление средних нормальное.

```python
alpha = .05 #критический уровень статистической значимости
results = st.ttest_ind(action, sports, equal_var=False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
print('Средний рейтинг Action', action.mean())
print('Средний рейтинг Sports', sports.mean())
```

Вывод: средние рейтинги жанра Action и Sport дейстительно разные.
